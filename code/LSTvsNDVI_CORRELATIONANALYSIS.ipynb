```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from rasterio.plot import show
year = 2017

# Load LST and NDVI data
lst_path = './CleanData/LST2015TRY.tif'
ndvi_path = './CleanData/NDVI2015TRY.tif'
with rasterio.open(lst_path) as lst_src:
    lst_data = lst_src.read(1)
    lst_max = np.nanmax(lst_data)
    lst_inv = lst_max - lst_data
with rasterio.open(ndvi_path) as ndvi_src:
    ndvi_data = ndvi_src.read(1)

lst_data = lst_inv
```

```python
# Calculate correlation coefficient
corr_coef = np.corrcoef(lst_data.flatten(), ndvi_data.flatten())[0][1]

# Print correlation coefficient
print(f"Correlation coefficient: {corr_coef}")

# Create map showing LST and NDVI data and correlation coefficient
fig, ax = plt.subplots(figsize=(10, 10))
show(lst_data, ax=ax, cmap='coolwarm', alpha=0.5)
show(ndvi_data, ax=ax, cmap='YlGn', alpha=0.5)
ax.set(title=f"Correlation Coefficient: {corr_coef:.2f}")
plt.show()
```
